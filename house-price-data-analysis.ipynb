## Missing Values

```python
# find the missing vales percentage in the whole datset

# 1. find all the features (columns) with missing values
feature_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() > 1]
print(f"features with missing values (out of 81):\n{feature_with_na}\n")
# 2. find the percentage of missing values in each one of feature and print it
print(f"percentage of missing values in each feature (column):\n")
for feature in feature_with_na:
    print(f"{feature}: {np.round(dataset[feature].isnull().mean(), 4)} %missing values")
```
